```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
import pywt
from scipy import misc
from PIL import Image
```

```python
def show_wavelet_decomp(img, level=1):
    coeffs = pywt.wavedec2(data=img, wavelet='db4', level=level)
    img,_ = pywt.coeffs_to_array(coeffs)
    plt.imshow(img, cmap='gray')
    return img
```

```python
ns = np.random.rand(100,100)
smear1 = np.eye(5)
smear2 = np.flip(np.eye(5), axis=0)
smear = smear1+smear2
```

```python
plt.figure()
plt.imshow(ns,cmap='gray')
plt.figure()
plt.imshow(smear1+smear2,cmap='gray')
```

```python
ns_smear = ndimage.convolve(ns, smear, mode='constant', cval=0.0)
```

```python
plt.imshow(ns_smear, cmap='gray')
```

```python
wv = show_wavelet_decomp(ns_smear, level=1)
```
